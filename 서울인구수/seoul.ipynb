# 3. Data visualization

```python
df['동별'].unique()
```

```python
df_set = df[df['동별'] =='서초구']
df_set = df_set.drop(columns=['동별','구분별'])
df_set = df_set.T.reset_index()
df_set =df_set.rename(columns={
    df_set.columns[0]:'year', 
    df_set.columns[1]: 'household',
    df_set.columns[2]: 'total_population',
    df_set.columns[3]: 'men',
    df_set.columns[4]: 'women',
    df_set.columns[5]: 'total_korean_population',
    df_set.columns[6]: 'korean_men',
    df_set.columns[7]: 'korean_women',
    df_set.columns[8]: 'total_foreigners',
    df_set.columns[9]: 'foreign_men',
    df_set.columns[10]: 'foreign_women',
    df_set.columns[11]: 'people_per_household',
    df_set.columns[12]: '65_years_old_and_up'
    })
```

```python
df_set_household = df_set[['year', 'household']]
df_set_household.plot(x = 'year', title='household')
plt.show()
```

```python
df_set_남녀_total =df_set[['year', 'total_population','men','women']]
df_set_남녀_total.plot(x = 'year', title='total_men&women')
plt.show()
```

```python
df_set_한국인_남녀_total =df_set[['year', 'total_korean_population','korean_men','korean_women']]
df_set_한국인_남녀_total.plot(x = 'year', title='total_korean_men&women')
plt.show()
```

```python
df_set_외국인_남녀_total = df_set[['year', 'total_foreigners','foreign_men','foreign_women']]
df_set_외국인_남녀_total.plot(x = 'year', title='total_foreigners_men&women')
plt.show()
```

```python
df_set_세대당인구 = df_set[['year', 'people_per_household']]
df_set_세대당인구.plot(x = 'year',title='people_per_household')
plt.show()
```

```python
df_set_65 = df_set[['year', '65_years_old_and_up']]
df_set_65.plot(x = 'year', title='65_years_old_and_up')
plt.show()
```
